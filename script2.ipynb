```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
%matplotlib nbagg
```

```python
ejeX = np.arange(-2*np.pi,2*np.pi,0.1)
vecI = np.sin(ejeX)
vecV = np.cos(ejeX)
vecP = vecI * vecV
```

```python
plt.figure()
plt.subplot(311)
plt.plot(vecI)
plt.title('vector I')
plt.grid(True)
plt.subplot(312)
plt.plot(vecV)
plt.title('vector V')
plt.grid(True)
plt.subplot(313)
plt.plot(vecP)
plt.title('vector P')
plt.grid(True)
p = np.sum(vecP)
print ("potencia total: " + str(p))
```
