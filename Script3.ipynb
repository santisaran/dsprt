```python
import numpy as np
import scipy as sp
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
%matplotlib nbagg
```

Señal de amplitud 1,fs = 10KHz, fsig = 1KHz, 1000 muestras

```python
#  Señal de amplitud 1, fs = 10kHz, fsig = 1kHz, 1000 muestras
N = 1000
Ts = 0.1 * 1e-3
fs = 1./Ts
fsig = 1000.
t = np.arange(0,N*Ts,Ts)
f = np.arange(0,fs,fs/(N))
Amp = 1
x = Amp * np.sin(2*np.pi*t*fsig);
espX = np.abs(sp.fft(x))

plt.figure()
plt.subplot(211)
plt.plot(t,x);
plt.title('Tiempo')
plt.grid(True)

plt.subplot(212)
plt.stem(f,espX);
plt.title('Frecuencia')
plt.grid(True)
```
